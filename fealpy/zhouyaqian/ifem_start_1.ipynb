```python
import numpy as np
from fealpy.mesh import TriangleMesh
import matplotlib.pyplot as plt # 画图
```

```python
node=np.array([(0,0),(1,0),(1,1),(0,1)])
cell=np.array([(0,1,2),(3,0,2)])

mesh=TriangleMesh(node,cell)
mesh.uniform_refine(n=2)
```

```python
fig=plt.figure(figsize=(10,10))
axes=fig.gca()
mesh.add_plot(axes)
mesh.find_node(axes,showindex=True,fontsize=20)
mesh.find_edge(axes,showindex=True,fontsize=20)
mesh.find_cell(axes,showindex=True)
plt.show()
```

```python
print('node:\n',node)
# print('edge:\n',edge)
print('cell:\n',cell)
print('mesh:\n',mesh)
```

```python
node=mesh.entity('node')
edge=mesh.entity('edge')
cell=mesh.entity('cell')
print('node:\n',node)
print('edge:\n',edge)
print('cell:\n',cell)
```

```python
print('geo_dimension top_dimenson',mesh.geo_dimension(),mesh.top_dimension()) # 获取网格的几何维数|拓扑维数
print('number_of_cells number_of_edges number_of_nodes',\
    mesh.number_of_cells(),mesh.number_of_edges(),mesh.number_of_nodes()) # 网格的节点个数|单元个数|边个数
# print('number of entities',mesh.number_of_entities('node'),mesh.number_of_entities('edge'),mesh.number_of_entities('cell'))

# display(mesh.integrator(0)) # 
print('entity_measure',mesh.entity_measure('node'),mesh.entity_measure('edge'),mesh.entity_measure('cell')) # 获取etype类型的实体的测度
print('entity_barycenter',mesh.entity_barycenter('cell'),mesh.entity_barycenter('edge')) # 获取etype类型的实体的重心
```

```python
cell2cell=mesh.ds.cell_to_cell() 
cell2edge=mesh.ds.cell_to_edge()
cell2node=mesh.ds.cell_to_node()
edge2cell=mesh.ds.edge_to_cell()
edge2edge=mesh.ds.edge_to_edge()
edge2node=mesh.ds.edge_to_node()
node2cell=mesh.ds.node_to_cell()
node2edge=mesh.ds.node_to_edge()
node2node=mesh.ds.node_to_node()

object_name=['cell2cell','cell2edge','cell2node','edge2cell','edge2edge','edge2node','node2cell','node2edge','node2node']
object_=[cell2cell,cell2edge,cell2node,edge2cell,edge2edge,edge2node,node2cell,node2edge,node2node]

for index in range(len(object_name)):
    print(object_name[index])
    for i,var in enumerate(object_[index]):
        print(i,':',var)
```
