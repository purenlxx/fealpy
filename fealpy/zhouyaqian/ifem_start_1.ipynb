### space and dof

```python
import sys
import numpy as np
import matplotlib.pyplot as plt

from fealpy.mesh import MeshFactory as mf
from fealpy.functionspace import LagrangeFiniteElementSpace
```

```python
p=3

box=[0,1,0,1]
mesh=mf.boxmesh2d(box,nx=2,ny=2,meshtype='tri')

space=LagrangeFiniteElementSpace(mesh,p)

ldof=space.number_of_local_dofs()
gdof=space.number_of_global_dofs()

bc=np.array([[1,0,0],[0,1,0],[0,0,1]])
phi=space.basis(bc)
gphi=space.grad_basis(bc)
```

```python
fig=plt.figure(figsize=(10,10))
axes=fig.gca()
mesh.add_plot(axes)
mesh.find_node(axes,showindex=True,fontsize=20)
mesh.find_edge(axes,showindex=True,fontsize=20)
mesh.find_cell(axes,showindex=True)
plt.show()
```

```python
print('ldof:',ldof)
print('gdof:',gdof)
print('bc:\n',bc)
print('phi:\n',phi.shape,phi)
print('gphi:\n',gphi.shape,gphi)
```

```python
ipoints=space.interpolation_points()
cell2dof=space.cell_to_dof() # 

for i,var in enumerate(cell2dof):
        print(i,':',var)
```

```python
##  
fig=plt.figure()
axes=fig.gca()
mesh.add_plot(axes)
mesh.find_node(axes,node=ipoints,showindex=True,color='r',fontsize=20)
plt.show()
```
